**Challenge 5** : Assume you are looking to build a model to predict if the text you are seeing is an issue or informational message. For e.g. job completed is an informational message whereas job abended is an issue. The objective is to predict the issue.

The task is multi fold

1. Generate the label for each data point to see if its informational or an issue
2. Build a count based matrix for each issue which reflects the number of occurrence of each word in the text along with the label
3. You can choose to restrict to a certain number of words based on your analysis 
4. You can choose to remove the STOPWORDS from each text 
5. The count based matrix that you build for the entire data set should be in a format such that it can be given as an input to any of the algorithm for training and testing.

#Import the libraries required

```
#Import the libraries required
import numpy as np
import pandas as pd
import nltk
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
nltk.download('punkt')
nltk.download('stopwords')
```

#Read Input file & Label creation

```
#Read Input file
data = pd.read_csv('ticket_Data.csv')
data.shape
type(data)

#Remove missing values
data = data.dropna()
# Removing the column TicketId. 
data = data.drop("TicketId", axis=1) 
#Lower text 
data['Description'] = data['Description'].str.lower() 
#searching values in description
search_values = ['completed','re-started','success','resolve','re-run','cleared']
# creating and passsing series to new column 
data["Indexes"]= data["Description"].str.contains('|'.join(search_values )) 
data['Indexes'].groupby(data['Indexes']).count() #Count of Indexes records
# creating label column 
data['label'] = np.where(data['Indexes']==False, 'Issue', 'Information')
data['label'].groupby(data['label']).count() #Count of issue vs Information records


```

# Data Preprocessing

```
# create selected column a list to enable further clean-up
data_list = list(data['Description'])
type(data_list)
#custom function
def tok(text):
  tokens = word_tokenize(text)
  return tokens
import regex as re
#Custom Preprocessing
def custom_preprocessor(text):
  #lowering the text case
  text = text.lower() 
  # remove special chars
  text = re.sub("\\W"," ",text)
  return text

cv2 = CountVectorizer(data_list,stop_words='english',min_df=2,tokenizer=tok,preprocessor=custom_preprocessor,lowercase=True,  token_pattern='(?u)\\b\\w\\w+\\b', ngram_range=(1, 4), analyzer='word')

#Lets test cv2 on our doc
cv2_doc = cv2.fit_transform(data_list)
cv2_doc
```

```
#@title Default title text
print(cv2.vocabulary_)
```
